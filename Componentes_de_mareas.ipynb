Para generar una se침al cuya amplitud est칠 modulada:

$$c(t) = A_c cos (洧랪_ct)$$ se침al original (portadora)

$$m(t) = A_m cos(洧랪_m t) $$ se침al moduladora

$$AM(t) = [A_c + m(t)] cos(洧랪_c t) $$ Se침al de Amplitud Modulada

```python
#Ejemplo de clase de modulaci칩n de amplitud
A_c = 1#float(input('Enter carrier amplitude: '))
f_c = 1/12#float(input('Enter carrier frquency: '))
A_m = 1#float(input('Enter message amplitude: '))
f_m = 1/(12*30)#float(input('Enter message frquency: '))
modulation_index = A_m/A_c#float(input('Enter modulation index: '))
```

```python
#t = np.linspace(0,30,24*30)
t=np.arange(0,30*24,0.5)
carrier = A_c*np.cos(2*np.pi*f_c*t)
modulator = A_m*np.cos(2*np.pi*f_m*t)
product = A_c*(1+modulation_index*np.cos(2*np.pi*f_m*t))*np.cos(2*np.pi*f_c*t)
```

```python
plt.subplot(3,1,1)
plt.title('Amplitude Modulation')
plt.plot(t/24,modulator,'g')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')

plt.subplot(3,1,2)
plt.plot(t/24,carrier, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')

plt.subplot(3,1,3)
plt.plot(product, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('AM signal')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(16, 9)
```
